```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score,accuracy_score
import pickle
```

```python
nltk.download("stopwords")
```

```python
dataset = pd.read_csv('datasets/reviews.txt',sep = '\t', names =['Reviews','Comments'])
```

```python
dataset
```

```python
stopset = set(stopwords.words('english'))
```

```python
vectorizer = CountVectorizer(stop_words=stopset)
```

```python
X = vectorizer.fit_transform(dataset.Comments)
y = dataset.Reviews
pickle.dump(vectorizer, open('transform.pkl', 'wb'))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X_train,y_train)
```

```python
accuracy_score(y_test,clf.predict(X_test))*100
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X,y)
```

```python
accuracy_score(y_test,clf.predict(X_test))*100
```

```python
clf.predict(vectorizer.transform(["This is one of the worst product I have ever used"]))
```

```python
filename = 'nlp_model.pkl'
pickle.dump(clf, open(filename, 'wb'))
```
